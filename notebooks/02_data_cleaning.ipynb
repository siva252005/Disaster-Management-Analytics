```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/1900_2021_disasters.csv")
```

```python
important_cols = [
    'Year',
    'Disaster Type',
    'Country',
    'Region',
    'Continent',
    'Total Deaths',
    'No Injured',
    'No Affected',
    'Total Affected',
    "Total Damages ('000 US$)"
]

df = df[important_cols]
```

```python
df.rename(columns={
    'Disaster Type':'Disaster_Type',
    'Total Deaths':'Deaths',
    'No Injured':'Injured',
    'No Affected':'Affected',
    'Total Affected':'Total_Affected',
    "Total Damages ('000 US$)":"Total_Damages_USD"
}, inplace=True)
```

```python
df.isnull().sum()
```

```python
print("Before:", df.shape)

df.drop_duplicates(inplace=True)

print("After:", df.shape)
```

```python
df.duplicated().sum()
```

```python
cols = [
    'Deaths',
    'Injured',
    'Affected',
    'Total_Affected',
    'Total_Damages_USD'
]

df[cols] = df[cols].fillna(0)
```

```python
df.isnull().sum()
```

```python
num_cols = [
    'Deaths',
    'Injured',
    'Affected',
    'Total_Affected',
    'Total_Damages_USD'
]

df[num_cols] = df[num_cols].astype(int)
```

```python
df.info()
```

```python
df['Decade'] = (df['Year']//10)*10
```

```python
def severity(deaths):
    if deaths >= 1000:
        return "High"
    elif deaths >= 100:
        return "Medium"
    else:
        return "Low"

df['Severity'] = df['Deaths'].apply(severity)
```
